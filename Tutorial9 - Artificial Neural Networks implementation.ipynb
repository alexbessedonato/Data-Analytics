```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

# load the dataset
file = 'pima-indians-diabetes.csv'
#header has to be 9 because dataframe has to be completely data
#columns go as follows
# 1. Number of times pregnant
# 2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
# 3. Diastolic blood pressure (mm Hg)
# 4. Triceps skin fold thickness (mm)
# 5. 2-Hour serum insulin (mu U/ml)
# 6. Body mass index (weight in kg/(height in m)^2)
# 7. Diabetes pedigree function
# 8. Age (years)
# 9. Class variable (0 or 1)
df = pd.read_csv(file, header=9)
# split into input (X) and output (y) variables
```

```python
df
```

```python
#X = first 8 columns
X = df.iloc[:,0:8]
#y = last column [1 or 0]
y = df.iloc[:,8]
```

```python
X
```

```python
y
```

```python
# We are going to train the entire dataset
# Normally one would split the data into training and test sets
knn = KNeighborsClassifier(n_neighbors=6, metric='euclidean')
knn.fit(X, y)
y_pred = knn.predict(X)
print("Mean hits: {:5.3f}".format(np.mean(y_pred==y)))
print("Accuracy: {:5.3f}".format(knn.score(X, y)))
# summarize the first 5 cases
for i in range(5):
	print(' predicted %d (actual %d)' % (y_pred[i], y[i]))
```

```python
# first neural network with keras make predictions
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense

# define the keras model
for count in range(10):
    model = Sequential()
    model.add(Dense(12, input_dim=8, activation='relu'))
    model.add(Dense(8, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    # compile the keras model
    model.compile(loss='binary_crossentropy', \
                  optimizer='adam', metrics=['accuracy'])

    # fit the keras model on the dataset
    model.fit(X, y, epochs = 150, batch_size=10, verbose=0)

    # make class predictions with the model
    predicted = model.predict(X)
    predicted=np.round(predicted).astype(int)

    _, accuracy = model.evaluate(X, y)
    print("Accuracy: {:5.3f}".format(accuracy))
    # summarize the first 5 cases
    for i in range(100):
        print(' predicted %d (actual %d)' % (predicted[i], y[i]))
```
